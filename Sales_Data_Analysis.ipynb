 ## Overall Sales trend

```python
sales.columns
```

```python
sales['Order Date']
```

```python
sales['Order Date'].min()
```

```python
sales['Order_Date']=pd.to_datetime(sales['Order Date'])
sales.info()
```

```python
sales.drop(columns=['Order Date'],inplace =True)
```

```python
sales['Month_Year']=sales['Order_Date'].dt.strftime ('%Y-%m')
```

```python
df_sales_monthly=sales.groupby('Month_Year').Sales.sum().reset_index()
df_sales_monthly
```

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation='85',size=9)
plt.plot(df_sales_monthly['Month_Year'],df_sales_monthly['Sales'],color='red')
plt.show()
```

# Which is the top selling product?

```python
sales.info()
```

```python
sales1=sales.loc[:,['Sales','Product Name']]
sales1
```

```python
prod_sales=sales1.groupby('Product Name').Sales.sum().reset_index().sort_values(by='Sales',ascending=False)
prod_sales
```

```python
#top ten selling products
prod_sales.head(10)
```

```python
#which product sold most quantities
sales2=sales.loc[:,['Product Name','Quantity','Sales']]
sales2
```

```python
sales2.groupby('Product Name')[['Quantity','Sales']].sum().reset_index().sort_values('Quantity',ascending=False)
```

```python
#top ten products selling maximum quantities
sales2.groupby('Product Name')[['Quantity','Sales']].sum().reset_index().sort_values('Quantity',ascending=False).head(10)
```
