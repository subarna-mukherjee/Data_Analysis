```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker=['ADBE','BA','KO']
```

```python
stocks=yf.download(ticker,start='2014-03-01',end='2020-09-01')
```

```python
stocks.info()
```

```python
stocks.to_csv('Stocks2.csv')
```

```python
stocks=pd.read_csv('Stocks2.csv',index_col=[0],header=[0,1],parse_dates=[0])
```

```python
stocks.head()
```

```python
stocks.columns=stocks.columns.to_flat_index()
stocks.head()
```

```python
stocks.columns=pd.MultiIndex.from_tuples(stocks.columns)
stocks.head()
```

```python
stocks.swaplevel(axis=1).sort_index(axis=1)
stocks
```

```python
Close=stocks.loc[:,'Close']
```

```python
Close.head()
```

```python
Close.div(Close.iloc[0]).mul(100)
```

```python
Close.plot(figsize=(15,8))
```

```python
stocks.head()
```

```python
Adobe=Close.ADBE.to_frame()
Adobe
```

```python
ret_adb=Adobe.pct_change(periods=1).dropna()
ret_adb
```

```python
ret_adb.plot(kind='hist',figsize=(15,10),bins=100,fontsize=15)
```

```python
Adobe.describe()
```

```python
mean_ret=ret_adb.mean()
mean_ret
```

```python
var_mean_ret=ret_adb.var()
var_mean_ret
```

```python
ret_adb.std()
```
