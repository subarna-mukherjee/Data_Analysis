```python
# to find which year had max num of courses introduced
df['Year']=df.published_timestamp.dt.year
```

```python
#to display how many  courses per subject,were introduced each year
df.groupby('Year').subject.value_counts()
```

```python
df.head(1)
```

```python
#find the year when maximum number of courses were introduced
total_course_year=df.groupby('Year').course_title.count().to_frame()
total_course_year.sort_values(by='course_title',ascending=False,inplace=True)
#In 2016,the maximum number of courses were launched
```

```python
total_course_year.reset_index()
```

```python
#by visualizing, we also reach the same conclusion,thta in 2016, the max number of courses were launched
sns.countplot(data=df,x='Year')
plt.show()
```
