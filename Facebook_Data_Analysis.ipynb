```python
fb.dropna(inplace=True)
fb.gender.isna().sum()
```

```python
#to find which gender is more active on fb through visualization
sns.countplot(x='gender',data=fb)
plt.show()
#as is evident, the males are more active on facebook
```

```python
fb['age_bracket']=pd.cut(fb.age,bins=[0,10,15,20,30,40,50,60,70,80,90,100])
fb.head()
```

```python
#now to find which age group is most active
fb.age_bracket.value_counts()
#as is evident maximum number of users lie between the age 20 to 30
print('Most active users lie in the age bracket:',fb.age_bracket.value_counts().head(1))
```

```python
#to find which gender is most active in which age group
plt.figure(figsize=(15,10))
sns.countplot(x='age_bracket',data=fb,hue='gender')
#as is evident in almost every age group,the male are more active. Except in age group, 70-80, where the females are little more active.
```

```python
fb.sample(n=5,random_state=123)
```

```python
#which gender gives more likes
fb.groupby('gender')[['mobile_likes','www_likes']].sum()
#Definitely females are more generous at giving likes, be it on the desktop or mobile versions
```

```python
fb.columns
```

```python
#which gender receives more likes
fb.groupby('gender')[['mobile_likes_received','www_likes_received']].sum()
#as we see females receive higher number of likes
```

```python
#does longer tenure increase the number of likes
fb.tenure.value_counts()
fb.groupby('tenure')[['likes_received']].sum().sort_values(by='likes_received',ascending=False)
fb.likes_received.corr(fb.tenure)
#Clearly there is no such rule to get likes
```
